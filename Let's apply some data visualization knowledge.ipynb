```python
def plot_correlation_map( TitanicP ):

    corr = TitanicP.corr()

    s , ax = plt.pyplot.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map( TitanicP )
```

The heat map is useful because it allows us to efficiently grasp which variables are strongly correlated with each other.

A co-efficient close to 1 means that there’s a very strong positive correlation between the two variables. In our case, the maroon shows very strong correlations. The diagonal line is the correlation of the variables to themselves — so they’ll obviously be 1.
I see a strong correlation between Sex and Survived for example.
