# Association Rules 

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
# Support Code
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
# Support with Column Names
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

60 % of transactions contain Hat and Skirt.
80 % of transactions contain Pants and Scraf.
60 % of transactions contain Pants and Scraf together and Pants and Skirt together.

```python
# Confidence Code
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

75 % of the transactions containing Pants have a Scraf and .
75 % of the transactions containing Pants have a Skirt.
100 % of the transactions containing Skirt have a Pants.

```python
# Lift Code
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

Lift (Skirt, Pants) = 1.25 > 1:  it is more likely to buy Skirt and Pants rather than Skirt alone.

# 2nd data set 

1. Data Preprocessing

```python
import pandas as pd
```

```python
data=pd.read_csv("Market_Basket_Optimisation.csv", header=None)

data.head()
```

```python
data.info()
```

```python
# replace all the NaN values with ‘’ and use inplace=True to commit the changes permanent into the dataframe
data.fillna('',axis=1,inplace=True)
data.head()
```

```python
# convert the dataframe into a list of lists where each inner list represents a transaction.

df_list = data.to_numpy().tolist()
df_list
dataset = list()
for i in range(len(df_list)) :
    item = list()
    for j in df_list[i] :
        if pd.notna(j):
            item.append(j)
    dataset.append(item)
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
# remove the first column as it does not contain any information
df.drop(columns=[''],axis=1,inplace=True)
# Each row in the dataframe represents a transaction.
df
```

```python
2. Applying Apriori Alogrithm
```

```python
# Support with Column Names
frequent_itemsets=apriori(df, min_support=0.01, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
# Confidence Code
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2) # associate itemsets with confidence over 20%.
```

```python
# Lift Code
from mlxtend.frequent_patterns import association_rules 
rules_ap = association_rules(frequent_itemsets,metric="lift",min_threshold=1.5)
```

```python
# Convert the rules obtained into a dataframe for better visualization
result = pd.DataFrame(rules_ap)
result.sort_values(by='lift',inplace=True,ascending=False)
result
```

Some observations from the final result :

it is more likely to buy herb and pepper and Ground beef rather than Ground beef alone. in addition Ground beef is 3.2 times more likely to be bought along with herb and pepper than by itself

People who buy spaghetti and mineral water often buy ground beef along with it.

People who buy chocolates also buy milk together rather than just buying them separately.
